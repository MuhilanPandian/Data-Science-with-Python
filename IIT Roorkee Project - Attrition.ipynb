### Problem Statement:

An Organization wants to find the likely candidates which are about to leave the organization.
    
 1. Build a supervised predictive model which are identify the likely candidates who are about to leave and can also predict the Probability as well so the company can design a preventive strategy
    
2. Model should be tested using the various evaluation metrics. Accuracy of the model should be high but also the accuracy
    
3. Category No should also be high as this is worry factor for the organization
    
4. Model should be tested before and after removing the class imbalance and impact of class imbalance should be taken into consideration
    
5. Feature selection should be performed and order of Importance along with the features should be tabulated. This feature selection will help the organization in decision making
    
6. Model should be created using more than one Algorithm and then the best algorithm should be chosen based on the business suitability
    
7. Required graphs should be plotted to see the performance for both the classesÂ

### Rearranging Columns

```python
column_titles = ['EmployeeNumber', 'Age', 'BusinessTravel', 'DailyRate', 'Department',
       'DistanceFromHome', 'Education', 'EducationField', 
        'EnvironmentSatisfaction', 'Gender', 'HourlyRate',
       'JobInvolvement', 'JobLevel', 'JobRole', 'JobSatisfaction',
       'MaritalStatus', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked',
       'Over18', 'OverTime', 'PercentSalaryHike', 'PerformanceRating',
       'RelationshipSatisfaction', 'StandardHours', 'StockOptionLevel',
       'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance',
       'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
       'YearsWithCurrManager', 'EmployeeCount', 'Attrition']

Attrition = Attrition.reindex(columns=column_titles)
Attrition.head(2)
```

### Splitting the data into Input variable and Output variable

```python
x = Attrition.iloc[:,1:-2] #selecting all the input columns leaving out Employee Number and Employee Count
y = Attrition.iloc[:,-1:]
```

```python
y.head(1)
```

### Splitting the data into Train and Test

```python
from sklearn.model_selection import train_test_split
```

## Building Models

## 1. Logistic Regression:

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size = 0.10, random_state=50)
```

```python
#Testing the shape
print(x_train.shape)
print(y_train.shape)
print('-----------')
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
Log_Reg = LogisticRegression()
```

```python
Log_Reg.fit(x_train,y_train)
```

### Prediction

```python
y_predict = Log_Reg.predict(x_test)
y_predict
#prediction on test data
#since the attrition with No or 0 Class is more in the data set, the prediction gave the output as 0
```

#### Checking the Performance of both the Classes

```python
sns.set_style('darkgrid') 
sns.countplot(y_predict)
```

### Evaluation Metrics

#### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_predict, y_test))
```

#### Accuracy, Precision, Recall, F1-Score

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_predict,y_test))
```

###### Even though accuracy seems to be good, the model performance was poor as there are no proper classifications made for Class 1 ('Yes')

## Probability

```python
y_predict_prob = pd.DataFrame(Log_Reg.predict_proba(x_test))
y_predict_prob.rename(columns={0 : "Class No", 1: "Class Yes"}, inplace=True)
```

###### Based on the below DataFrame, the company can formulate the preventive strategy for those whose probability for Class Yes is high

```python
Probability = pd.concat([Attrition.EmployeeNumber, pd.Series(y_predict_prob['Class No']), pd.Series(y_predict_prob['Class Yes'])], axis = 1)
Probability.head(10)
```

### ROC & AUC score and graph

```python
from sklearn.metrics   import roc_curve, roc_auc_score
```

```python
AUC = roc_auc_score(y_test , y_predict_prob.iloc[ : , 1])
AUC
#gives the AUC score
```

```python
FPR , TPR  , Threshold = roc_curve( y_test ,  y_predict_prob.iloc[ : , 1])
```

```python
plt.figure(figsize=  (8,8))
plt .plot(FPR, TPR, label="AUC="+str(AUC))
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC CURVE")
plt.grid(b=True)
plt.legend(loc=4)
```

###### Because of imbalance in the data, even a slightest increase in TPR would significantly affect the FPR also beyond 0.39
