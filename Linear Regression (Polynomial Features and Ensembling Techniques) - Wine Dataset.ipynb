# Ensembling:

# Bagging

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split (x,y, test_size=0.1, random_state=0)
```

```python
from sklearn.ensemble import BaggingRegressor
bag_reg = BaggingRegressor(n_estimators = 20, random_state = 0,verbose=2)
bag_reg.fit(Xtrain,Ytrain)
```

```python
bag_dash = bag_reg.predict(Xtest)
bag_dash
```

```python
Ytest
```

```python
Bagpredict= bag_reg.predict(np.array([[7.8,0.880,0.0,2.6,0.098,25.0,67.0,0.99680,3.20,0.68]]))
Bagpredict
```

```python
wine_1
```

```python
R2_Bag = r2_score(Ytest, bag_dash)
R2_Bag*100
```

# Random Bagging

```python
from sklearn.ensemble import RandomForestRegressor
ranfor_reg = RandomForestRegressor(n_estimators = 20, random_state = 0,verbose=2)
ranfor_reg.fit(Xtrain,Ytrain)
```

```python
ranfor_dash = ranfor_reg.predict(Xtest)
ranfor_dash
```

```python
R2_ranfor = r2_score(Ytest,ranfor_dash)
R2_ranfor*100
```

```python
ranfor_predict = ranfor_reg.predict(np.array([[7.8,0.880,0.0,2.6,0.098,25.0,67.0,0.99680,3.20,0.68]]))
ranfor_predict
```

# ExtraTrees Bagging

```python
from sklearn.ensemble import ExtraTreesRegressor
ext_reg = ExtraTreesRegressor(n_estimators=10, random_state=0, verbose=0)
ext_reg.fit(Xtrain,Ytrain)
```

```python
ext_dash = ext_reg.predict(Xtest)
ext_dash
```
