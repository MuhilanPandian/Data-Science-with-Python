```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv ('health_insurance_cost.csv')
data
```

```python
data.describe()
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
data.sex=LE.fit_transform(data.sex)
data.smoker=LE.fit_transform(data.smoker)
```

```python
data.info()
```

```python
data['sex']=data['sex'].astype('category')
data['smoker']=data['smoker'].astype('category')
```

```python
data.head()
```

```python
data.describe()
```

```python
data1 = data.drop('region',axis=1)
data1
```

```python
x=data1.iloc[:,:-1]
x.head()
```

```python
y=data1.iloc[:,-1:]
y.head()
```

# Spliting the Data

```python
from sklearn.model_selection import train_test_split
XTrain, XTest, Ytrain, Ytest = train_test_split (x,y, test_size= 0.1, random_state=1 )
```

```python
XTrain
```

# Training

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit (XTrain, Ytrain)
```

# Testing

```python
ydash= LR.predict(XTest)
ydash
```

```python
Ytest
```

```python
LR.coef_
```

```python
LR.intercept_
```

```python
from sklearn.metrics import r2_score,mean_squared_error
R2 = r2_score(Ytest, ydash)
```

```python
R2*100
```

# Mean Squared Error (MSE)

```python
MSE = mean_squared_error(Ytest, ydash)
MSE
```

# Root Mean Squared Error (RSME)

```python
RSME = np.sqrt(mean_squared_error(Ytest, ydash))
RSME
```
